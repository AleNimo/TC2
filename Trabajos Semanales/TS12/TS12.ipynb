## Punto 1
### Simulación simbólica de $Y_{11}$

```python
import sympy as sp
import splane as tc2
import schemdraw
from schemdraw import Drawing
from schemdraw.elements import Capacitor
from schemdraw.elements import Inductor
from schemdraw.elements import Resistor

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
Y11 = 3*s*(s**2+7/3)/((s**2+2)*(s**2+5))

#Y21 tiene ceros en el origen, j1 e infinito ==> Al sintentizar Y11 debo remover polos solamente en esas frecuencias.

#Remoción parcial de impedancia en el origen para mover cero de jsqrt(2) a j1:
Z11 = 1/Y11

Z11_A, ZA, C1 = tc2.remover_polo_dc(Z11, omega_zero = 1)

#Remoción total de polo en j1 en admitancia

Y11_A = 1/Z11_A

Y11_B, YB, L1, C2 = tc2.remover_polo_jw(Y11_A, omega = 1, isImpedance = False)

#Remoción total de polo en infinito en impedancia

Z11_B = 1/Y11_B

Z11_C, ZC, L2 = tc2.remover_polo_infinito(Z11_B)

#La impedancia resultante es de un capacitor

C3 = 1/(Z11_C*s)

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V1$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y_{11}',  
                                          Y11, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_serie(d, Capacitor, C1)

d = tc2.dibujar_tanque_derivacion(d, L1, C2)
    
d = tc2.dibujar_elemento_serie(d, Inductor, L2)

d = tc2.dibujar_elemento_serie(d, Capacitor, C3)
                        
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_puerto_salida(d, 
                        voltage_lbl = ('+', '$V2$', '-'), 
                        current_lbl = '$I2$')
display(d)
```

### Simulación circuital de parámetros:

### $Y_{11}=\frac{I_1}{V_1} \big| _{V_2=0}$

### $Y_{21}=\frac{I_2}{V_1} \big| _{V_2=0}$

#### Circuito simulado:



#### Se observa como se cortocircuita el puerto 2 de manera tal que $V_2=0$

#### Se comparan los parametros simulados con las expresiónes planteadas en el enunciados:

### $Y_{11}=\frac{3S\left(S^2+\frac{7}{3}\right)}{(S^2+2)(S^2+5)}$

### $Y_{21}=\frac{S\left(S^2+1\right)}{(S^2+2)(S^2+5)}$

```python
import scipy.signal as sig
import ltspice
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18})

l = ltspice.Ltspice("./Simulaciones Circuitales/Punto1.raw")
l.parse()

fig, axes = plt.subplots(4, 1, figsize=(15,15))

#Obtengo curvas de simulación en ltspice

freq = l.get_frequency()
Y11_lt = -l.get_data('I(C1)') / l.get_data('V(V1)')   #Y11 = I1/V1 (V2=0)
Y21_lt = l.get_data('I(C3)') / l.get_data('V(V1)')    #Y21 = I2/V1 (V2=0)

#Comparo con la función de excitación de la consigna
Y11_num = [3,0,7,0]
Y11_den = [1,0,7,0,10]
w, Y11 = sig.freqs(Y11_num,Y11_den,freq*2*np.pi)

Y21_num = [1,0,1,0]
Y21_den = [1,0,7,0,10]
w, Y21 = sig.freqs(Y21_num,Y21_den,freq*2*np.pi)

axes[0].set_title("Módulo de Y11")
axes[0].semilogx(freq*2*np.pi, 20*np.log10(np.abs(Y11_lt)), linewidth=3.0, label='LTspice')
axes[0].semilogx(w, 20*np.log10(np.abs(Y11)), 'r--', linewidth=3.0, label='Analítico')

axes[1].set_title("Fase de Y11")
axes[1].semilogx(freq*2*np.pi, np.angle(Y11_lt), linewidth=3.0, label='LTspice')
axes[1].semilogx(w, np.angle(Y11), 'r--', linewidth=3.0, label='Analítico')

axes[2].set_title("Módulo de Y21")
axes[2].semilogx(freq*2*np.pi, 20*np.log10(np.abs(Y21_lt)), linewidth=3.0, label='LTspice')
axes[2].semilogx(w, 20*np.log10(np.abs(Y21)), 'r--', linewidth=3.0, label='Analítico')

axes[3].set_title("Fase de Y21")
axes[3].semilogx(freq*2*np.pi, np.angle(Y21_lt), linewidth=3.0, label='LTspice')
axes[3].semilogx(w, np.angle(Y21), 'r--', linewidth=3.0, label='Analítico')

axes[0].grid()
axes[1].grid()
axes[2].grid()
axes[3].grid()

axes[1].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')
axes[3].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes[0].set_ylabel("Amplitude [dB]")
axes[1].set_ylabel("Phase [rad]")

axes[2].set_ylabel("Amplitude [dB]")
axes[3].set_ylabel("Phase [rad]")

axes[0].legend()
axes[1].legend()
axes[2].legend()
axes[3].legend()

plt.tight_layout()

plt.show() 
```

#### Se observa que la curva analítica y la simulada con LTspice coinciden, por lo que se verifica correctamente que el circuito calculado cumple con los parámetros de admitancia solicitados.

## Punto 2 - Parámetros Z

### Simulación simbólica de $Z_{11}$

```python
s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
Z11 = ((s+2)*(s+4))/((s+1)*(s+3))

#Al sintentizar Z11 debo remover polos solamente en -1 e infinito.

#Remoción total polo en -1 (impedancia):

Z11_A, ZA, R1, C1 = tc2.remover_polo_sigma(Z11, sigma = 1, isImpedance=True, isRC=True)

#Remoción total de constante en infitito (impedancia):

Z11_B, R2 = tc2.remover_valor_en_infinito(Z11_A, isImpedance = True)

#Remoción total de constante en origen (admitancia):

Y11_C, R3 = tc2.remover_valor_en_dc(1/Z11_B, isImpedance = False)

#La admitancia resultante es de un capacitor (remoción de polo en infinito):

C2 = Y11_C/s

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V1$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z_{11}',  
                                          Z11, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_tanque_RC_serie(d, R1, C1)

d = tc2.dibujar_elemento_serie(d, Resistor, R2)

d = tc2.dibujar_elemento_derivacion(d, Resistor, R3)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_puerto_salida(d, 
                        voltage_lbl = ('+', '$V2$', '-'), 
                        current_lbl = '$I2$')
display(d)
```

## Punto 2 - Parámetros Y

### Simulación simbólica de $Y_{22}$

```python
s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
Y22 = ((s+2)*(s+4))/(s+3)

#Al sintentizar Y22 debo remover polos solamente en -1 e infinito.

#Remoción parcial de constante en el origen (admitancia) para para forzar un cero en -1:

Y22_A, R1 = tc2.remover_valor(Y22, sigma_zero = 1, isImpedance=False)

#Remoción total de polo en -1 (impedancia):

Z22_B, ZA, R2, C1 = tc2.remover_polo_sigma(1/Y22_A, sigma = 1, isImpedance=True, isRC=True)

#Remoción total de polo en infinito (admitancia):

Y22_C, YB, C2 = tc2.remover_polo_infinito(1/Z22_B)

#La impedancia resultante es de un resistor:

R3 = 1/Y22_C

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V1$', '-'), 
                        current_lbl = '$I$')

d = tc2.dibujar_elemento_serie(d, Resistor, R3)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_tanque_RC_serie(d, R2, C1)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Resistor, R1)

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y_{22}',  
                                          Y22, 
                                          hacia_salida = False,
                                          k_gap_width = 0.5)

d = tc2.dibujar_puerto_salida(d, 
                        voltage_lbl = ('+', '$V2$', '-'), 
                        current_lbl = '$I2$')
display(d)
```

### Simulación Circuital de ambas redes contrastándolas las transferencias obtenidas:

#### Usando parámetros Z: $T_{(S)} = \frac{V_2}{V_1} \big | _{I_2=0} = \frac{1}{2} \frac{(S+1)}{(S+2)(S+4)}$

#### Usando parámetros Y: $T_{(S)} = \frac{V_2}{V_1} \big | _{I_2=0} = \frac{7}{2} \frac{(S+1)}{(S+2)(S+4)}$

#### Circuitos simulados:



```python
l = ltspice.Ltspice("./Simulaciones Circuitales/Punto2.raw")
l.parse()

fig, axes = plt.subplots(4, 1, figsize=(15,15))

#Obtengo curvas de simulación en ltspice

freq = l.get_frequency()
T_Z_lt = l.get_data('V(V2_Z)')  #(V1=1)
T_Y_lt = l.get_data('V(V2_Y)')  #(V1=1)

#Comparo con las funciones transferencias obtenidas analíticamente
num_z = [0.5,0.5]
num_y = [3.5, 3.5]
den = [1,6,8]

w, T_Z = sig.freqs(num_z,den,freq*2*np.pi)
w, T_Y = sig.freqs(num_y,den,freq*2*np.pi)
    
axes[0].set_title("Módulo de T(s) (parámetros Z)")
axes[0].semilogx(freq*2*np.pi, 20*np.log10(np.abs(T_Z_lt)), linewidth=3.0, label='LTspice')
axes[0].semilogx(w, 20*np.log10(np.abs(T_Z)), 'r--', linewidth=3.0, label='Analítico')

axes[1].set_title("Fase de T(s) (parámetros Z)")
axes[1].semilogx(freq*2*np.pi, np.angle(T_Z_lt), linewidth=3.0, label='LTspice')
axes[1].semilogx(w, np.angle(T_Z), 'r--', linewidth=3.0, label='Analítico')

axes[2].set_title("Módulo de T(s) (parámetros Y)")
axes[2].semilogx(freq*2*np.pi, 20*np.log10(np.abs(T_Y_lt)), linewidth=3.0, label='LTspice')
axes[2].semilogx(w, 20*np.log10(np.abs(T_Y)), 'r--', linewidth=3.0, label='Analítico')

axes[3].set_title("Fase de T(s) (parámetros Y)")
axes[3].semilogx(freq*2*np.pi, np.angle(T_Y_lt), linewidth=3.0, label='LTspice')
axes[3].semilogx(w, np.angle(T_Y), 'r--', linewidth=3.0, label='Analítico')

axes[0].grid()
axes[1].grid()
axes[2].grid()
axes[3].grid()

axes[1].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')
axes[3].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes[0].set_ylabel("Amplitude [dB]")
axes[1].set_ylabel("Phase [rad]")

axes[2].set_ylabel("Amplitude [dB]")
axes[3].set_ylabel("Phase [rad]")

axes[0].legend()
axes[1].legend()
axes[2].legend()
axes[3].legend()

plt.tight_layout()

plt.show() 
```

#### En ambas redes circuitales se observa que la curva analítica y la simulada con LTspice coinciden, por lo que se verifica correctamente que los circuitos calculados cumplen con la transferencia solicitada.
