## Simulación circuital de parámetros S:

### Circuito simulado:



### Se compara la simulación con las expresiones calculadas analíticamente de los parámetros:

### $S_{11} = \frac{S^3+S(1-R_o^2)}{S^3+2R_oS^2+S(1+R_o^2)+R_o}$

#### (Eligiendo $R_o = 1 = R_g = R_L$)

```python
import scipy.signal as sig
import ltspice
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18})

l = ltspice.Ltspice("./Simulaciones Circuitales/Punto1.raw")
l.parse()

#Obtengo parámetros S de simulación en ltspice
freq = l.get_frequency()
S11_spice = l.get_data('S11(v1)')
S21_spice = l.get_data('S21(v1)')

#Comparo con los parámetros obtenidos analíticamente
Ro = 1

num_S11 = [1, 0, (1-Ro**2), 0]
den_S11 = [1, 2*Ro, (1+Ro**2), Ro]

num_S21 = [Ro]
den_S21 = [1, 2*Ro, (1+Ro**2), Ro]

w, S21 = sig.freqs(num_S21,den_S21,freq*2*np.pi)
w, S11 = sig.freqs(num_S11,den_S11,freq*2*np.pi)

fig, axes = plt.subplots(4, 1, figsize=(15,15))
axes[0].set_title("Módulo de S11")
axes[0].semilogx(freq*2*np.pi, 20*np.log10(np.abs(S11_spice)), linewidth=3.0, label='LTspice')
axes[0].semilogx(w, 20*np.log10(np.abs(S11)), 'r--', linewidth=3.0, label='Analítico')

axes[1].set_title("Fase de S11")
axes[1].semilogx(freq*2*np.pi, np.angle(S11_spice), linewidth=3.0, label='LTspice')
axes[1].semilogx(w, np.angle(S11), 'r--', linewidth=3.0, label='Analítico')

axes[2].set_title("Módulo de S21")
axes[2].semilogx(freq*2*np.pi, 20*np.log10(np.abs(S21_spice)), linewidth=3.0, label='LTspice')
axes[2].semilogx(w, 20*np.log10(np.abs(S21)), 'r--', linewidth=3.0, label='Analítico')

axes[3].set_title("Fase de S21")
axes[3].semilogx(freq*2*np.pi, np.angle(S21_spice), linewidth=3.0, label='LTspice')
axes[3].semilogx(w, np.angle(S21), 'r--', linewidth=3.0, label='Analítico')

axes[0].grid()
axes[1].grid()
axes[2].grid()
axes[3].grid()

axes[1].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')
axes[3].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes[0].set_ylabel("Amplitude [dB]")
axes[1].set_ylabel("Phase [rad]")
axes[2].set_ylabel("Amplitude [dB]")
axes[3].set_ylabel("Phase [rad]")

axes[0].legend()
axes[1].legend()
axes[2].legend()
axes[3].legend()

plt.tight_layout()

plt.show() 
```

#### Se observa que ambos parámetros coinciden con los simulados. Por otro lado al ser una red pasiva y simétrica: $S_{12} = S_{21}$ y $S_{11} = S_{22}$

## a) Analizando comportamiento de la red a partir de parámetro $S_{21}$:

### A partir de la fórmula: $S_{21} = \frac{V_2}{\frac{V_{g1}}{2}}$

### Se puede deducir que $\frac{S_{21}}{2} = \frac{V_2}{V_{g1}}$ que resulta ser la transferencia de tensión de la red T estando doblemente cargada con $R_g = R_L = 1$ (Valores normalizados)

### Entonces:

## $\frac{S_{21}}{2} = \frac{V_2}{V_{g1}} = \frac{1}{2} \frac{1}{S^3+2S^2+2S+1} = \frac{1}{2} \frac{1}{(S+1)(S^2+S+1)}$

### Resulta ser un filtro pasa bajos Butterworth de tercer orden con $\omega_{corte} = 1$ (También normalizada)

### b) Analizando el comportamiento de la red en frecuencias puntuales a partir de parámetros $S_{11}$ y $S_{21}$:

* $\omega = 0$
* $\omega = 1 (\omega_{corte})$
* $\omega \rightarrow \infty$

#### Analíticamente se llegó a los siguientes valores:

| $$\omega$$ |  $$S_{11_{(\omega)}}$$ |  $$S_{21_{(\omega)}}$$ |
|:----------:|:----------------------:|:----------------------:|
|      0     |            0           |            1           |
|      1     | $$\frac{1}{\sqrt{2}}$$ | $$\frac{1}{\sqrt{2}}$$ |
| $$\infty$$ |            1           |            0           |

#### A continuación se superponen las curvas de $S_{11}$ y $S_{21}$ en un gráfico, mostrando el punto de $\omega = 1$

```python
fig, axes = plt.subplots(1, 1, figsize=(15,5))
axes.set_title(r"Módulo de $S_{11}$ y $S_{21}$")
axes.semilogx(freq*2*np.pi, 20*np.log10(np.abs(S11_spice)), linewidth=3.0, label=r'$S_{11}$')
axes.semilogx(w, 20*np.log10(np.abs(S21_spice)), 'r', linewidth=3.0, label=r'$S_{21}$')

axes.plot(1, 20*np.log10(1/np.sqrt(2)), 'go', markersize=12, label = r'$|S_{11_{(1)}}| = |S_{21_{(1)}}| = \frac{1}{\sqrt{2}}$')

axes.grid()

axes.set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes.set_ylabel("Amplitude [dB]")

axes.legend()

plt.show()
```

* ### En $\omega = 0$:
    
    ##### $|S_{11}|^2$ (Coef. de relfexión en puerto 1) es máximo y $|S_{21}|^2$ (Coef. de transferencia directa) es nulo por lo que toda la potencia de la fuente es reflejada (nada pasa a la carga en el puerto 2).
* ### En $\omega = 1$:
    
    ##### $|S_{11}|^2 = |S_{21}|^2 = \frac{1}{2}$ lo que quiere decir que una mitad de la potencia de la fuente es reflejada, y la otra mitad es transmitida a la carga.
* ### En $\omega \rightarrow \infty$:
    
    ##### $|S_{11}|^2$ es nulo y $|S_{21}|^2$ es máximo por lo que toda la potencia de la fuente es transferida a la carga.
