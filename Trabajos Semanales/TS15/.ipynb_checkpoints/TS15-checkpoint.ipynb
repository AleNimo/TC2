### Luego obtengo la red operando simbólicamente con $Z_1$

```python
import sys
sys.path.insert(1, 'D:/UTN/Nivel 4/Teoría de los circuitos 2/Repositorio/TC2')  #para poder usar un solo splane.py y no tener que copiar el archivo coef todas las carpetas
import splane as tc2

import sympy as sp
import schemdraw
from schemdraw import Drawing
from schemdraw.elements import Capacitor
from schemdraw.elements import Inductor
from schemdraw.elements import Resistor

s = sp.symbols('s ', complex=True)

# Z1 calculada con ayuda de matlab:
Z1 = (2.099051337347753 * s**4 + s**3 + 2.575457027007341 * s**2 + 0.779398094130057 * s + 0.420468930901884)\
/(s**3 + 0.476405689659571 * s**2 + 0.779398094130047 * s + 0.158087513733408)

#La síntesis debe generar una red escalera por lo que procedo coef remover solo en infinito:

ZA,K1,L1 = tc2.remover_polo_infinito(Z1)

# Luego de cada remoción debo redondear los coeficientes para que la remoción sea total (sino quedan coeficientes espuria)
for coef in sp.preorder_traversal(ZA):
    if isinstance(coef, sp.Float):
        ZA = ZA.subs(coef, round(coef, 14))

YB,K2,C1 = tc2.remover_polo_infinito(1/ZA)

for coef in sp.preorder_traversal(YB):
    if isinstance(coef, sp.Float):
        YB = YB.subs(coef, round(coef, 12))

ZC,K3,L2 = tc2.remover_polo_infinito(1/YB)

for coef in sp.preorder_traversal(ZC):
    if isinstance(coef, sp.Float):
        ZC = ZC.subs(coef, round(coef, 11))

YD,K4,C2 = tc2.remover_polo_infinito(1/ZC)

YD = YD.subs(s,0)

# La admitancia constante resultante es la resistencia de carga
RL = 1/YD
# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V_1$', '-'), 
                        current_lbl = '$I_1$')

# Redondeo un poco Z1 para mostrarlo en el circuito:
Z1_redondeado = Z1
for coef in sp.preorder_traversal(Z1):
    if isinstance(coef, sp.Float):
        Z1_redondeado = Z1_redondeado.subs(coef, round(coef, 5))

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z_1',  
                                          Z1_redondeado, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)


d = tc2.dibujar_elemento_serie(d, Inductor, L1)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d = tc2.dibujar_elemento_serie(d, Inductor, L2)

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_puerto_salida(d, 
                        voltage_lbl = ('+', '$V_2$', '-'), 
                        current_lbl = '$I_2$')

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Resistor, round(RL, 5))

display(d)
```

### Se compara el valor de |S21|^2 de la simulación con la transferencia Chebyshev de la que se parte en el diseño

```python
import scipy.signal as sig
import ltspice
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18})

l = ltspice.Ltspice("./Simulaciones Circuitales/Normalizado.raw")
l.parse()

#Obtengo parámetros S con ltspice
freq = l.get_frequency()
S11_ltspice = l.get_data('S11(v1)')
S21_ltspice = l.get_data('S21(v1)')

#Comparo con |S21|^2 de Chebyshev

w = freq*2*np.pi

ep_2 = 10**0.1 - 1

S21_2 = 1/(1 + ep_2 * (8*w**4-8*w**2+1)**2) # |S21(w)|^2

fig, axes = plt.subplots(1, 1, figsize=(15,7))
axes.set_title(r'$|S_{21_{(\omega)}}|^2$')
axes.semilogx(w, 20*np.log10(np.abs(S21_ltspice)**2), linewidth=3.0, label='LTspice')
axes.semilogx(w, 20*np.log10(S21_2), 'r--', linewidth=3.0, label=r'Chebyshev $4^{to}$ Orden')

axes.grid()

axes.set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes.set_ylabel("Amplitud [dB]")

axes.legend()

axes.set_ylim(-3, 1)
axes.set_xlim(w[0], 1.5)

plt.tight_layout()

plt.show() 
```

#### Vemos que el parametro $|S_{21}|^2$ de la red coincide con la transferencia de un Chebyshev de cuarto orden.
#### (El ripple es el doble de la plantilla dado que $S_{21}$ está elevado al cuadrado).

### Ahora se grafican $S_{11}$ y $S_{21}$ como pide la consigna:

```python
fig, axes = plt.subplots(2, 1, figsize=(15,10))
axes[0].set_title(r'$|S_{21}|$ (LTspice)')
axes[0].semilogx(w, 20*np.log10(np.abs(S21_ltspice)), 'r', linewidth=3.0)
axes[0].set_ylim(-2,1)

axes[0].axvline(x = 0.383,ymin=-0.5, color = 'k', ls = '--', clip_on=False)
axes[0].plot(0.383, 0, marker="o", markersize=7, markeredgecolor="blue", markerfacecolor="blue")

axes[0].axvline(x = 0.924,ymin=-0.5, color = 'k', ls = '--', clip_on=False)
axes[0].plot(0.924, 0, marker="o", markersize=7, markeredgecolor="blue", markerfacecolor="blue")

axes[1].set_title(r'$|S_{11}|$ (LTspice)')
axes[1].semilogx(w, 20*np.log10(np.abs(S11_ltspice)), linewidth=3.0)
axes[1].axvline(x = 0.383, color = 'k', ls = '--')
axes[1].axvline(x = 0.924, color = 'k', ls = '--')

# axes[2].set_title(r'$\angle S_{11}$ y $\angle S_{21}$')
# axes[2].semilogx(w, np.unwrap(np.angle(S11_ltspice)), linewidth=3.0, label=r'$\angle S_{11}$')
# axes[2].semilogx(w, np.unwrap(np.angle(S21_ltspice)), 'r', linewidth=3.0, label=r'$\angle S_{21}$')

axes[0].set_xlim(w[0], 1.5)
axes[1].set_xlim(w[0], 1.5)
# axes[2].set_xlim(w[0], 1.5)

axes[0].grid()
axes[1].grid()
# axes[2].grid()

axes[1].set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

axes[0].set_ylabel("Amplitud [dB]")
axes[1].set_ylabel("Fase [rad]")

# axes[2].legend()

plt.tight_layout()

plt.show() 
```

#### El ripple de $S_{21}$ respeta el solicitado de 1dB. La frecuencia de corte es $\omega = 1$ por lo que el filtro está correctamente normalizado.
#### Se observa que cuando la transferencia directa de potencia es máxima: $\left( |S_{21}| = 1 \rightarrow |S_{21}|^2 = 1 \right)$, el coeficiente de reflexión de potencia en el puerto 1 es mínimo: $\left( |S_{11}| = 0 \rightarrow |S_{11}|^2 = 0 \right)$ y viceversa.

#### Esta relación es evidente desde el punto de vista físico, y se hace evidente matemáticamente cuando se analiza la ecuación de la que se parte la síntesis de la red no disipativa:

#### $ |S_{11}|^2 = 1 - |S_{21}|^2$

### Comportamiento de la red a partir del valor de $S_{11}$

#### (Se utiliza $|S_{11}|^2$ que tiene una interpretación física directa)

```python
fig, axes = plt.subplots(1, 1, figsize=(15,5))
axes.set_title(r'$|S_{11}|^2$ (LTspice)')
axes.semilogx(w, np.abs(S11_ltspice)**2, linewidth=3.0)
axes.axvline(x = 1, color = 'k', ls = '--')
axes.axvline(x = 1.88, color = 'k', ls = '--')
axes.grid()

axes.set_ylabel("Amplitud (lineal)")
axes.set_xlabel(r'$\omega \left[\frac{rad}{seg}\right]$')

plt.tight_layout()

plt.show() 
```

* En el centro de la banda de paso se refleja el 20% de la potencia que llega al puerto 1.
* En la frecuencia de corte nuevamente se refleja el 20% (está relacionado a la atenuación máxima en banda de paso $\alpha_{max} = 10\cdot log(1+\epsilon^2)$ ).
* La transición ocurre entre $\omega = 1$ y $\omega \approx 1.88$ donde la reflexión alcanza prácticamente el 100% y se mantiene así por el resto de la banda de rechazo.

#### De esta manera queda evidenciado el comportamiento de filtro pasa bajos, permitiendo el 80% de la potencia que llega al puerto 1 en la mitad de la banda de paso, y reflejando toda la energía luego de transicionar a la banda de rechazo.

## Desnormalización del circuito
### $f_{corte} = 10^6$ y resistencia del generador de $50\Omega$

```python
# Norma de frecuencia:
norm_w = 2*np.pi*10**6 #(w_corte)

# Norma de impedancia:
norm_z = 50 #(RG)

# Desnormalización de componentes:
# R = R'* norm_z
# L = L'* norm_z /norm_w
# C = C'/(norm_w*norm_z)
RL_desnorm = RL * norm_z

L1_desnorm = L1 * norm_z / norm_w
L2_desnorm = L2 * norm_z / norm_w

C1_desnorm = C1 / (norm_z * norm_w)
C2_desnorm = C2 / (norm_z * norm_w)

# Dibujamos la red resultante:

d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V_1$', '-'), 
                        current_lbl = '$I_1$')

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_serie(d, Inductor, L1_desnorm)

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1_desnorm)

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_serie(d, Inductor, L2_desnorm)

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2_desnorm)

d = tc2.dibujar_puerto_salida(d, 
                        voltage_lbl = ('+', '$V_2$', '-'), 
                        current_lbl = '$I_2$')

d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)
d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Resistor, round(RL_desnorm, 5))

display(d)
```
