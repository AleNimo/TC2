## Simulacin Circuital de Pasa-Bajos y Pasa-Banda

### $\omega_0=1$;$Q = 3$;$|T(0)| = 20dB = K$





### Se puede observar como tienen la misma $\omega_0$

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import signal as sig # mdulo de SciPy

# un mdulo adaptado a mis necesidades
from splane import analyze_sys, pretty_print_lti

# Para que el tamao de la salida sea un poco ms adecuada al tamao del documento
mpl.rcParams['figure.figsize'] = (10,10)

# Estilo de los grficos
##############################################################
fig_font_family = 'Ubuntu'
fig_font_size = 16

plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})

npoints = 1000
##############################################################

#Grfico del pasabajos

R1 = 10e3
R2 = 300e3
R3 = 100e3
R4 = 100e3
C = 10e-6

K = -R3/R1
W0 = 1/(C * R3)
Q = R2/R3

num = K*W0**2
den = np.array([1, W0/Q, W0**2])

sistema = sig.TransferFunction(num,den)

pretty_print_lti(sistema)

analyze_sys(sistema,-3,7) #Funcion cambiada para poder determinar frec inicial y final de grficos
```

```python
#Grfico del pasa-banda

K2 = -R2/R1

num2 = np.array([K2 * W0/Q, 0])
den2 = np.array([1, W0/Q, W0**2])

sistema2 = sig.TransferFunction(num2,den2)

pretty_print_lti(sistema2)

analyze_sys(sistema2,-3,7) #Funcion cambiada para poder determinar frec inicial y final de grficos
```
