## Simulación Numérica Transferencia Normalizada

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import signal as sig # módulo de SciPy

# un módulo adaptado a mis necesidades
import splane as tc2

# Para que el tamaño de la salida sea un poco más adecuada al tamaño del documento
mpl.rcParams['figure.figsize'] = (10,10)

# Estilo de los gráficos
##############################################################
fig_font_family = 'Ubuntu'
fig_font_size = 16

plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})

npoints = 1000
##############################################################

Wp=1
Wz=1/3
Qp=1

num1 = np.array([1,0,Wz**2])
den1 = np.array([1, Wp/Qp, Wp**2])

num2 = np.array([1,0])
den2 = np.array([1, Wp**2])

H = sig.TransferFunction(np.convolve(num1, num2),np.convolve(den1, den2))

#Muestro la expresión de la transferencia
print("\nTransferencia H(s):")
tc2.pretty_print_lti(H)

#La muestro separada en transferencias de segundo y primer orden
#(No uso tc2.tf2sos_analog y luego pretty_print_SOS porque distribuye distinto las constantes)
H1 = sig.TransferFunction(num1,den1)
H2 = sig.TransferFunction(num2,den2)
print("\nSeparandola en sistemas SOS:")
print("\nH1(s):")
tc2.pretty_print_lti(H1)
print("\nH2(s):")
tc2.pretty_print_lti(H2)



tc2.analyze_sys(H,-3,3)
```
